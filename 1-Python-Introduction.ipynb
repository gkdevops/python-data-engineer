```python
my_list = [1,2,3,4,5]

def reverse_list(list):
    start = 0
    end = len(list) - 1
    
    while start < end:
        list[start], list[end] = list[end], list[start]
        start += 1
        end -= 1
    return list

rev_list = reverse_list(my_list)

print (rev_list)
```
